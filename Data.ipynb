```python
import matplotlib.pyplot as plt
import pandas as pd
from pandas_profiling import ProfileReport
import seaborn as sns
%matplotlib inline
```

```python
dat = pd.read_excel('NAC 3.0 - Dataset.xlsx')
dat.drop(dat.columns[[1]], axis=1, inplace=True)
dat_tr = dat.T
headers = dat_tr.iloc[0]
dat_trf = pd.DataFrame(dat_tr.values[1:], columns=headers)
dat_trf = dat_trf.astype('float64')
```

```python
corr = dat_trf.corr()
plt.figure(figsize=(14,8))
sns.set_theme(style="white")
heatmap = sns.heatmap(corr, cmap="Blues", fmt='.1g')
```
